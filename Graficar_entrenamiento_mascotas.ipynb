```python
def parse_line(l):
  splitted = l.split("\t")
  stripped = [l.strip(",") for l in splitted]
  second_values = [stripped[0]] + [p.split(": ")[1] for p in stripped[1:]]
  almost_ready = [second_values[0]] + [second_values[1].split("/")[0]] + second_values[2:]
  return (almost_ready[0],) + (int(almost_ready[1]),) + tuple([float(p) for p in almost_ready[2:]])
a = [l.strip() for l in salida.split("\n")]
a = [l for l in a if len(l)]
a = [parse_line(l) for l in a]

datetimes = [l[0] for l in a]
epochs = [l[1] for l in a]
losses = [l[2] for l in a]
triplet_accs = [l[3] for l in a]
val_losses = [l[4] for l in a]
val_triplet_accs = [l[5] for l in a]

print(datetimes)
print(epochs)
print(losses)
print(triplet_accs)
print(val_losses)
print(val_triplet_accs)
```

```python
def blur_list(l, n=1):
  return l
  numbers = [ l[max(0, i-n) : i] for i in range(1, len(l) + 1) ]
  return [sum(x)/len(x) for x in numbers]

plt.rcParams["figure.figsize"] = (20, 7)

def make_plot(xss, ylabel):
  show_last = 0
  for (label, xs) in xss:
    plt.plot(epochs[-show_last:], blur_list(xs)[-show_last:], label=label)
  verticales = lambda valores, col: plt.vlines(valores, 0, max([max(xs[-show_last:]) for (label, xs) in xss]), linestyles='dashed', colors=col)
  verticales([249.5, 549.5], 'red') # interrupciones entrenamiento
  verticales([1092, 1103, 507, 793], 'green') # redes descargadas y subidas a github/mascotas-general
  plt.grid(axis='y')
  plt.xlabel('Epoch')
  plt.ylabel(ylabel)
  plt.title('Entrenamiento mascotas')
  plt.legend()
  plt.show()

make_plot([("Training", losses), ("Validation", val_losses)], 'Loss')
make_plot([("Training", triplet_accs), ("Validation", val_triplet_accs)], 'Accuracy')
```
